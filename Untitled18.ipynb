```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
print("Google Drive mounted successfully!")


# Define the file path for the dataset
file_path = '/content/drive/My Drive/fake_currency_data.csv'


df = pd.read_csv(file_path)
print("\nDataset loaded successfully!")
display(df.head())
```

```python

y = df['Counterfeit']
X = df.drop(['Counterfeit', 'SerialNumber'], axis=1)

print("\nFeatures (X) extracted:")
display(X.head())
print("\nTarget variable (y) extracted:")
display(y.head())

categorical_features = ['Country', 'Denomination', 'SecurityFeatures']
numerical_features = ['Weight', 'Length', 'Width', 'Thickness']


preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('scaler', StandardScaler(), numerical_features)
    ],
    remainder='passthrough'
)


print("\nApplying preprocessing (One-Hot Encoding and Scaling)...")
X_processed = preprocessor.fit_transform(X)
print("Preprocessing complete.")
print(f"Shape of processed features (X_processed): {X_processed.shape}")



X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.2, random_state=42, stratify=y
)

print(f"\nShape of training features (X_train): {X_train.shape}")
print(f"Shape of testing features (X_test): {X_test.shape}")
print(f"Shape of training target (y_train): {y_train.shape}")
print(f"Shape of testing target (y_test): {y_test.shape}")

# Check the distribution of the target variable in training and testing sets after splitting
print("\nTarget distribution in original data:")
display(y.value_counts(normalize=True))
print("\nTarget distribution in training set:")
display(y_train.value_counts(normalize=True))
print("\nTarget distribution in testing set:")
display(y_test.value_counts(normalize=True))
```

```python

print("\nInitializing and training the XGBoost model...")
xgb_model = xgb.XGBClassifier(objective='binary:logistic', random_state=42)

xgb_model.fit(X_train, y_train)
print("XGBoost model training complete.")
```

```python

y_pred = xgb_model.predict(X_test)

print("\nPredictions made on the testing data.")



accuracy = accuracy_score(y_test, y_pred)
print(f"\nModel Accuracy: {accuracy:}")


report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(report)


conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)


plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted Genuine (0)', 'Predicted Fake (1)'],
            yticklabels=['Actual Genuine (0)', 'Actual Fake (1)'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('XGBoost Confusion Matrix')
plt.show()


print("\nModel evaluation complete. The accuracy and classification report are shown above.")
```
